```python
df_net = netflix.copy()
df_net
```

```python
df_net.dropna(thresh=2,inplace=False)
df_net
```

```python
"""""
using fillna to replace nan values with unknown in cast , director and country
"""
df_net["cast"].fillna("Unkown Actor", inplace=True)
df_net["director"].fillna("Unkown director", inplace=True)
df_net["country"].fillna("Unkown country", inplace=True)

df_net.head(10)
```

```python
df_net['date_added'] = df_net['date_added'].str.strip()
```

```python

df_net['date_added'] = pd.to_datetime(df_net['date_added'], errors='coerce', infer_datetime_format=True)

df_net['day_added'] = df_net['date_added'].dt.day
df_net['year_added'] = df_net['date_added'].dt.year
df_net['month_added']=df_net['date_added'].dt.month
df_net
```

```python
""""
 Merging the new data set for cast and diretor on title
"""
x=pd.merge(cast_new , director_new , on="title" , how="inner")
x
```

```python
""""
 Merging the new data set for x and country on title
"""
y=pd.merge(x,country_new , on="title" , how="inner")
y
```

```python
""""
 Merging the new data set for y and listed in on title
"""
z=pd.merge(y , genre_new , on="title" , how="inner")
z
```

```python
""""
 Merging the new data set for z and copy of netflix data set  on title and creating a new cleaner data set
"""
final_net=pd.merge(z,df_net,on="title", how="inner" )
final_net
```

```python
final_net.drop(["director_y" ,"cast_y","country_y" ,"listed_in_y"] , axis=1 , inplace=True)
```
